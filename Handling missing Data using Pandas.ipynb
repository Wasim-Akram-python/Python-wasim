# Interpolate

Linear interpolation involves estimating a new value by connecting two adjacent known values with a straight line.
If the two known values are (x1, y1) and (x2, y2), then the y value for some point x is: Linear interpolation is a
    straight line fit between two data points.

```python
df.interpolate()
```

```python
df.interpolate(method='time')
```
